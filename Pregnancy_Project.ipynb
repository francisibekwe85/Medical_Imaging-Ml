### Handling Missing Feature Values

#### You CANNOT train a model with missing target values,  You must REMOVE those rows

```python
df = df.dropna(subset=['Risk Level'])
```

```python
df.shape
```

### Fill the rest with Median

```python
df['BMI'].fillna(df['BMI'].median(), inplace=True)
df['Systolic BP'].fillna(df['Systolic BP'].median(), inplace=True)
df['Diastolic'].fillna(df['Diastolic'].median(), inplace=True)
df['BS'].fillna(df['BS'].median(), inplace=True)
df['Heart Rate'].fillna(df['Heart Rate'].median(), inplace=True)

# Binary columns (0/1)
df['Previous Complications'].fillna(0, inplace=True)
df['Preexisting Diabetes'].fillna(0, inplace=True)
```

```python
df.isna().sum()
```

### Encode target

```python
df['Risk Level'] = df['Risk Level'].map({'Low':0, 'High':1})
```

#### "Missing values in the dataset were handled using median imputation for numerical features, while rows with missing target values were removed to ensure model integrity."

```python
from sklearn.model_selection import train_test_split

X = df.drop('Risk Level', axis=1)
y = df['Risk Level']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train.value_counts()
```

### Apply SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

### Check New Distribution

import pandas as pd

print(pd.Series(y_train_smote).value_counts())

### Train Model on SMOTE Data

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train_smote, y_train_smote)
```

###  Evaluate on REAL Test Data (Not SMOTE!)

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Generate Confusion Matrix (Table Format)

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

cm = confusion_matrix(y_test, y_pred)

cm_table = pd.DataFrame(
    cm,
    index=['Actual Low (0)', 'Actual High (1)'],
    columns=['Predicted Low (0)', 'Predicted High (1)']
)

cm_table
```

### Cross-validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, y, cv=5)
print(scores.mean())
```

### Build a Fully Deterministic Pipeline

### Fix ALL Randomness Globally

```python
import numpy as np
import random
import os

SEED = 42

np.random.seed(SEED)
random.seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
```

### Use Pipeline (CRITICAL)

👉 Never apply SMOTE manually again
👉 Put everything inside a pipeline

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier

pipeline = Pipeline([
    ("smote", SMOTE(random_state=SEED)),
    ("xgb", XGBClassifier(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=5,
        random_state=SEED,
        eval_metric='logloss',
        subsample=1.0,              # remove randomness
        colsample_bytree=1.0,       # remove randomness
        n_jobs=1                    # VERY IMPORTANT (stability)
    ))
])
```

### Stable Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=SEED,
    stratify=y
)
```

### Train Model (Through Pipeline Only)

```python
pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
```

### Evaluate

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

### Generate Classification Report (Table)

```python
from sklearn.metrics import classification_report
import pandas as pd

report = classification_report(
    y_test,
    y_pred,
    output_dict=True
)

report_df = pd.DataFrame(report).transpose().round(4)

report_df
```

#### 🟦 1. Abstract

This project focuses on predicting pregnancy risk levels (High or Low) using machine learning techniques based on clinical and demographic features. Two models — Random Forest and XGBoost — were developed and evaluated. Special emphasis was placed on model stability, reproducibility, and clinical safety, particularly minimizing false negatives (missed high-risk cases). After addressing class imbalance using SMOTE and ensuring deterministic model behavior through a pipeline approach, XGBoost emerged as the most reliable and deployable model.

### 🟦 2. Methodology
Data Handling
* Dataset contained clinical indicators such as blood pressure, blood sugar, BMI, and medical history.
* Missing values were handled appropriately.
* Target variable encoded as:
*  0 → Low Risk
*  1 → High Risk
Class Imbalance Handling
* SMOTE (Synthetic Minority Oversampling Technique) was applied.
* Implemented inside a pipeline to prevent data leakage and ensure consistency.




Model Development

Two models were trained:

* Random Forest Classifier
* XGBoost Classifier





Stability Enhancement (Critical Contribution)

To ensure reproducibility:

* Fixed random seeds (random_state = 42)
* SMOTE integrated inside pipeline
XGBoost configured with:
* subsample = 1.0
* colsample_bytree = 1.0
* n_jobs = 1

This ensured fully deterministic results across multiple runs.

### 3  Styled Random Forest Report

```python
rf_report_df = pd.DataFrame({
    "Class": ["Low Risk (0)", "High Risk (1)", "Accuracy"],
    "Precision": [0.99, 0.98, None],
    "Recall": [0.99, 0.99, None],
    "F1-Score": [0.99, 0.98, 0.99],
    "Support": [141, 97, 238]
})

def highlight_rf(val, column):
    if column == "F1-Score" and val == 0.99:
        return "background-color: lightgreen"
    return ""

rf_report_df.style.apply(
    lambda col: [highlight_rf(v, col.name) for v in col],
    axis=0
)
```

### Styled XGBoost Report (BEST MODEL)

```python
xgb_report_df = pd.DataFrame({
    "Class": ["Low Risk (0)", "High Risk (1)", "Accuracy"],
    "Precision": [1.0000, 0.9794, None],
    "Recall": [0.9860, 1.0000, None],
    "F1-Score": [0.9930, 0.9896, 0.9916],
    "Support": [143, 95, 238]
})

def highlight_xgb(val, column):
    if column in ["Precision", "Recall", "F1-Score"] and val is not None:
        if val == xgb_report_df[column].dropna().max():
            return "background-color: lightgreen"
    if column == "Recall" and val == 1.0:
        return "background-color: lightgreen; font-weight: bold"
    return ""

xgb_report_df.style.apply(
    lambda col: [highlight_xgb(v, col.name) for v in col],
    axis=0
)
```

### Styled Confusion Matrix (Most Important)
XGBoost (highlight safety)

```python
import pandas as pd

# Create XGBoost confusion matrix table
xgb_cm_df = pd.DataFrame(
    [[141, 2],
     [0, 95]],
    index=["Actual Low Risk (0)", "Actual High Risk (1)"],
    columns=["Predicted Low Risk (0)", "Predicted High Risk (1)"]
)

# Styling function
def highlight_cm(data):
    styles = pd.DataFrame("", index=data.index, columns=data.columns)

    # True Negative
    styles.iloc[0, 0] = "background-color: lightgreen; font-weight: bold"

    # False Positive
    styles.iloc[0, 1] = "background-color: khaki"

    # False Negative
    styles.iloc[1, 0] = "background-color: lightgreen; font-weight: bold"

    # True Positive
    styles.iloc[1, 1] = "background-color: lightgreen; font-weight: bold"

    return styles

# Display styled table
xgb_cm_df.style.apply(highlight_cm, axis=None)
```

### Styled Model Comparison Table

```python
import pandas as pd

# Create comparison table
comparison_df = pd.DataFrame({
    "Metric": [
        "Accuracy",
        "False Negatives",
        "False Positives",
        "Recall (High Risk)",
        "Stability",
        "Reproducibility"
    ],
    "Random Forest": [
        "99%",
        1,
        2,
        0.99,
        "Unstable",
        "Inconsistent"
    ],
    "XGBoost": [
        "99.16%",
        0,
        2,
        1.00,
        "Stable",
        "Consistent"
    ]
})

# Styling function
def highlight_comparison(val, metric):
    if metric == "Accuracy" and val == "99.16%":
        return "background-color: lightgreen"
    
    if metric == "False Negatives":
        if val == 0:
            return "background-color: lightgreen; font-weight: bold"
        else:
            return "background-color: red; color: white"
    
    if metric == "Recall (High Risk)" and val == 1.00:
        return "background-color: lightgreen; font-weight: bold"
    
    if metric == "Stability" and val == "Stable":
        return "background-color: lightgreen"
    
    if metric == "Reproducibility" and val == "Consistent":
        return "background-color: lightgreen"
    
    return ""

# Apply styling
comparison_df.style.apply(
    lambda row: [highlight_comparison(val, row["Metric"]) for val in row],
    axis=1
)
```

###  🟦 5. Discussion

Although both models achieved high accuracy, their behavior differed significantly in terms of reliability and clinical safety.

* Random Forest demonstrated strong performance but suffered from instability, producing slightly different results across runs. It also recorded one false negative, meaning a high-risk pregnancy could be missed.


* XGBoost, after applying a deterministic pipeline approach, produced consistent results across multiple runs. Most importantly, it achieved 100% recall for high-risk cases, ensuring that no high-risk pregnancy was misclassified.

  

In healthcare applications, false negatives are critical because failing to identify a high-risk patient can delay intervention and lead to severe consequences. Therefore, eliminating false negatives was prioritized over minimizing false positives.
